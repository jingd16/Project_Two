```python
# import dependencies 
import os
import pandas as pd
```

```python
# generate path to csv
csv_path = os.path.join('output', 'country.csv')
country_csv = pd.read_csv(csv_path)
country_csv.head()
country_csv=country_csv[["country", 
                        "prct_rpt_crime", 
                        "gdp",
                        "prct_leisure_satis_high",
                        "prct_life_satis_high",
                        "total_pop",
                        "police_trust_rating",
                        "prct_rpt_pollution",
                        "avg_temp",
                        "avg_precipitation"]]
```

```python
country_csv.head()
```

```python
country_csv.rename(columns={'country':'Country',
                          'gdp':'GDP',
                          'prct_rpt_crime':'Crime %',
                           "prct_leisure_satis_high": 'Leisure_Satis %',
                           'prct_life_satis_high' : "Life_Satis %",
                           "total_pop":"Population",
                           "police_trust_rating" : "Plice_Trust %",
                            "prct_rpt_pollution" : "Pollution %",
                            "avg_temp" : "Avg Temp °F",
                            "avg_precipitation" : "Avg Precipitation %"}, inplace=True)
```

```python
country_csv.head()
```

```python
# create html table and save to file 
country_csv.to_html('table.html', index=False, classes=['table', 'table-striped', 'table-hover'])
```
