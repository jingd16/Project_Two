```python
#Import Dependencies
import pandas as pd
from sqlalchemy import create_engine
from config import password
```

```python
#import all csv
crime_path = "data/crime_2016.csv"
gdp_path = "data/gdp_2016.csv"
leisure_path = "data/leisure_satisfaction_2013.csv"
life_sat_path = "data/life_satisfaction_2013.csv"
pollution_path = "data/pollution_2016.csv"
population_path = "data/population_2011.csv"
trust_police_path = "data/trust_in_police_2013.csv"
weather_path = "data/weather.csv"

#Import the CSV into a pandas DataFrame
crime_df = pd.read_csv(crime_path, low_memory = False)
gdp_df = pd.read_csv(gdp_path, low_memory = False)
leisure_df = pd.read_csv(leisure_path, low_memory = False)
life_sat_df = pd.read_csv(life_sat_path, low_memory = False)
pollution_df = pd.read_csv(pollution_path, low_memory = False)
population_df = pd.read_csv(population_path, low_memory = False)
trust_police_df = pd.read_csv(trust_police_path, low_memory = False)
weather_df = pd.read_csv(weather_path, low_memory = False)
```

```python
weather_df=weather_df[["country", "avg_temp", "avg_low_temp", "avg_precipitation"]]
weather_df.head(5)
```

```python
# ================merge all data frame (if required)================QUESTION?
country_df = pd.merge(crime_df, gdp_df, left_on="country", right_on="country", how='left')
country_df = pd.merge(country_df, leisure_df, left_on="country", right_on="country", how='left')
country_df = pd.merge(country_df, life_sat_df, left_on="country", right_on="country", how='left')
country_df = pd.merge(country_df, pollution_df, left_on="country", right_on="country", how='left')
country_df = pd.merge(country_df, population_df, left_on="country", right_on="country", how='left')
country_df = pd.merge(country_df, trust_police_df, left_on="country", right_on="country", how='left')
country_df = pd.merge(country_df, weather_df, left_on="country", right_on="country", how='left')
country_df.head(5)
```

```python
#export to csv file
country_df.to_csv("output/country.csv")
```
