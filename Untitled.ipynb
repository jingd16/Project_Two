```python
#Import Dependencies
import pandas as pd
from sqlalchemy import create_engine
from config import password
```

```python
world_cities = "data/worldcities.csv"
world_cities = pd.read_csv(world_cities)
world_cities.head()
```

```python
world_cities['city'] = world_cities['city'].str.lower()
world_cities['country']=world_cities['country'].str.lower()
world_cities=world_cities[["city","country","lat", "lng" ]]
world_cities.head(200)
```

```python
cities = "data/cities_final_urls.csv"
cities = pd.read_csv(cities)
cities.head()
```

```python
new_cities= pd.merge(cities, world_cities, how="left", left_on=["city","country"], right_on = ["city","country"])
```

```python
new_cities.to_csv("output/top_cities.csv")
```
