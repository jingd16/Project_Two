```python
#new URL
url_new = "https://www.roadaffair.com/best-places-to-visit-in-europe/"
executable_path = {'executable_path': ChromeDriverManager().install()}
browser_new = Browser('chrome', **executable_path, headless=False)
```

```python
browser_new.visit(url_new)

response = requests.get(url_new)
soup = BeautifulSoup(response.text, 'html.parser')
```

```python
print(soup.prettify())
```

```python
all_cities_urls = []

# HTML object
html_new = browser_new.html
# Parse HTML with Beautiful Soup
soup2 = BeautifulSoup(html_new, 'html.parser')
#data = soup2.find('div', class_='entry-content')
urls = soup2.find_all('figure', class_='wp-caption')

len(urls)
       # Iterate through each url
for urls in urls:
    city_name_url = {}
    city_name_url ['image_url']  = urls.find('img')['data-lazy-src']
    all_cities_urls.append(city_name_url)
        # Use Beautiful Soup's find() method to navigate and retrieve attributes
    print(city_name_url)
    
```

```python
print(all_cities_urls)
```

```python
url_df = DataFrame (all_cities_urls,columns=['image_url'])
```

```python
url_df.head()
```

```python
url_df['location'] = url_df['image_url']
```

```python
url_df.head()
```

```python
url_df['location'] = url_df['location'].str.replace("https://www.roadaffair.com/wp-content/uploads/2017/09/", "")
```

```python
url_df
```

```python
url_df['location'] = url_df['location'].str.split("-shutterstock", n = 1, expand = True)
```

```python
import_fields = ['restaurant_name','country','city','address','latitude','longitude',
                 'top_tags','price_range','meals','cuisines','special_diets','features',
                 'vegetarian_friendly','vegan_options','gluten_free','avg_rating',
                 'total_reviews_count','excellent','very_good','average','poor','terrible',
                 'food','service','value','atmosphere']

raw_restaurants_df= pd.read_csv('tripadvisor_european_restaurants.csv', usecols=import_fields, dtype={
                'restaurant_name': str,
                'country': str,
                'city': str,
                'address': str,
                'latitude': float,
                'longitude': float,
                'top_tags': str,
                'price_level': str,
                'price_range': str,
                'meals': str,
                'cuisines': str,
                'special_diets': str,
                'features': str,
                'vegetarian_friendly': str,
                'vegan_options': str,
                'gluten_free': str,
                'avg_rating': float,
                'total_reviews_count': float,
                'excellent': float,
                'very_good': float,
                'average': float,
                'poor': float,
                'terrible': float,
                'food': float,
                'service': float,
                'value': float,
                'atmosphere': float
                 })
```

```python
raw_restaurants_df
```

```python
reviewed_restaurants = raw_restaurants_df.loc[(raw_restaurants_df["total_reviews_count"] > 200)]
reviewed_restaurants
```

```python
located_restaurants = reviewed_restaurants.dropna(subset=['latitude'])
located_restaurants
```

```python
food_restaurants = located_restaurants.dropna(subset=['food'])
food_restaurants
```

```python
avg_rating = food_restaurants.dropna(subset=['avg_rating'])
avg_rating
```

```python
five__star_restaurants = avg_rating.loc[(avg_rating["avg_rating"] > 4.5)]
five__star_restaurants
```

```python
five__star_restaurants.to_csv("five__star_restaurants.csv")
```
